```python
from transformers import pipeline

# Hugging Face pipeline 使用簡單 API 呼叫模型
classifier = pipeline("sentiment-analysis")
```

```python
# 測試文字分類
result = classifier("I love using Hugging Face for AI applications!")
print(result)
```

```python
classifier = pipeline("sentiment-analysis", model="uer/roberta-base-finetuned-jd-binary-chinese")
result = classifier("這個模型非常好用！")
print(result)
```

```python
classifier = pipeline("sentiment-analysis", model="uer/roberta-base-finetuned-jd-binary-chinese")
result = classifier("這個模型也太爛了吧！")
print(result)
```

```python
import json

# 保存結果至 JSON 文件
with open("output.json", "w", encoding="utf-8") as f:
    json.dump(result, f, ensure_ascii=False, indent=4)
```
